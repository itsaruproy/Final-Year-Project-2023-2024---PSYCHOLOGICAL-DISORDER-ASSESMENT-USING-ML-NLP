```python
import pandas as pd
data = pd.read_csv("data.csv", delimiter='\t')
print(data.columns.values)
```

```python
df = data[[f'Q{i}A' for i in range(1, 43)]]
print(df.columns)
```

```python
df.columns = [f'Q{i}' for i in range(1, 43)]
df.head()
```

```python
df.shape
```

```python
df = df - 1
df.head()
```

```python
anxiety_df = df[['Q2', 'Q4', 'Q7', 'Q9', 'Q15', 'Q19', 'Q20', 'Q23', 'Q25', 'Q28', 'Q30', 'Q36', 'Q40', 'Q41']]
stress_df = df[['Q1', 'Q6', 'Q8', 'Q11', 'Q12', 'Q14', 'Q18', 'Q22', 'Q27', 'Q29', 'Q32', 'Q33', 'Q35', 'Q39']]
depression_df = df[['Q3', 'Q5', 'Q10', 'Q13', 'Q16', 'Q17', 'Q21', 'Q24', 'Q26', 'Q31', 'Q34', 'Q37', 'Q38', 'Q42']]
```

```python
anxiety_df.loc[:, 'WS'] = anxiety_df.sum(axis=1)
stress_df.loc[:, 'WS'] = stress_df.sum(axis=1)
depression_df.loc[:, 'WS'] = depression_df.sum(axis=1)
```

```python
anxiety_df.head()
```

```python
bins = [0, 8, 10, 15, 20, float('inf')]
labels = [0, 1, 2, 3, 4]
anxiety_df['Classes'] = pd.cut(anxiety_df['WS'], bins=bins, labels=labels, right=False)
anxiety_df.head()
```

```python
bins = [0, 15, 19, 26, 34, float('inf')]
labels = [0, 1, 2, 3, 4]
stress_df['Classes'] = pd.cut(stress_df['WS'], bins=bins, labels=labels, right=False)
stress_df.head()
```

```python
bins = [0, 10, 14, 21, 28, float('inf')]
labels = [0, 1, 2, 3, 4]
depression_df['Classes'] = pd.cut(depression_df['WS'], bins=bins, labels=labels, right=False)
depression_df.head()
```

```python
anxiety_df.drop('WS', axis=1, inplace=True)
stress_df.drop('WS', axis=1, inplace=True)
depression_df.drop('WS', axis=1, inplace=True)
```

```python
anxiety_df.to_csv('anxiety.csv', index=False)
stress_df.to_csv('stress.csv', index=False)
depression_df.to_csv('depression.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_selection import mutual_info_classif

X = anxiety_df.drop(labels=['Classes', 'WS'], axis=1)
y = anxiety_df['Classes']

info_gain = mutual_info_classif(X, y)

# Create a DataFrame to store feature names and their information gain scores
feature_info_gain = pd.DataFrame({'Feature': X.columns, 'Information Gain': info_gain})

# Sort features based on information gain
feature_info_gain = feature_info_gain.sort_values(by='Information Gain', ascending=False)

print(feature_info_gain)

# Plot the information gain scores
plt.figure(figsize=(10, 6))
plt.barh(feature_info_gain['Feature'], feature_info_gain['Information Gain'], color='skyblue')
plt.xlabel('Information Gain')
plt.title('Feature Importance (Information Gain)')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)
```

```python
class_distribution = y_train.value_counts()

# Display the class distribution
print("Class Distribution:")
print(class_distribution)
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
sampler = SMOTE()
X_train, y_train = sampler.fit_resample(X_train, y_train)
```

```python
class_distribution = y_train.value_counts()

# Display the class distribution
print("Class Distribution:")
print(class_distribution)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
dtclf = DecisionTreeClassifier(criterion='entropy')
dtclf.fit(X_train, y_train)
```

```python
y_pred = dtclf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score
print(classification_report(y_test, y_pred))
```
